```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import matplotlib.pyplot as plt
################## Plt style to see values in VSC
from matplotlib import style
style.use('dark_background')
##################
import time
```

# Full Code

```python
# Creating a driver to let python open and execute a chrome browser instance

options = Options()
options.page_load_strategy = 'eager'

driver = webdriver.Chrome(ChromeDriverManager().install(),options=options)

# Get the driver to the required link
page_url = 'https://lotr.fandom.com/wiki/Category:The_Lord_of_the_Rings_characters'
driver.get(page_url)

# Click the cookies Accept button (CHANGE TEXT FROM 'ACEPTAR' TO REQUIRED LANGUAGE IF NECESSARY)
time.sleep(3)
driver.find_element(By.XPATH, '//div[text()="ACEPTAR"]').click()

# Find All races categories and get their name and urls
characters = driver.find_elements(By.CLASS_NAME, "category-page__member-link")

character_list = []
for char in characters:
    char_name = char.text
    character_list.append({'char_name': char_name})
```

```python
character_list
```

```python
# let's remove 3 wrongly scrapped character elements because of these 'Categories' inside the webpage

character_list.remove({'char_name': 'Category:Characters that have appeared in the Hobbit and the Lord of the Rings'})

character_list.remove({'char_name': 'Category:Major characters (The Lord of the Rings)'})

character_list.remove({'char_name': 'Category:Minor characters (The Lord of the Rings)'})
```

```python
# Create a Pandas DataFrame from our character list
character_df = pd.DataFrame(character_list)
character_df
```

```python
# Finally, we should rename hobbit's names, because they are merely mentioned with their full names.

character_df.iloc[9]['char_name']='Merry'
character_df.iloc[19]['char_name']='Sam'
character_df.iloc[41]['char_name']='Pippin'
```
